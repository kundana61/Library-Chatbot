```python
library_data = {
    "intents": {
        "greeting": {
            "patterns": ["hi", "hello", "hey"],
            "responses": ["Welcome to the Library!", "Hello! How may I help you?"]
        },

        "membership": {
            "patterns": ["membership", "library card", "register"],
            "responses": ["Library membership is free for students."]
        },

        "issue_limit": {
            "patterns": ["how many books", "issue limit", "borrow limit"],
            "responses": ["Students can issue up to 3 books."]
        },

        "fine_info": {
            "patterns": ["fine", "late fine", "penalty"],
            "responses": ["Fine is ₹5 per day after due date."]
        },

        "library_rules": {
            "patterns": ["rules", "library rules"],
            "responses": ["Maintain silence and return books on time."]
        },

        "goodbye": {
            "patterns": ["bye", "exit"],
            "responses": ["Thank you! Visit again."]
        }
    },

    "books": {
        "data structures": 4,
        "machine learning": 2,
        "dbms": 5
    }
}
```

```python
import random

def detect_intent(user_input):
    user_input = user_input.lower()
    for intent, data in library_data["intents"].items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return intent
    return "unknown"

def check_availability(user_input):
    for book, copies in library_data["books"].items():
        if book in user_input:
            if copies > 0:
                return f"{book.title()} book is available ({copies} copies)."
            else:
                return f"{book.title()} book is currently unavailable."
    return "Book not found in library."

def calculate_fine(user_input):
    for word in user_input.split():
        if word.isdigit():
            days = int(word)
            fine = days * 5
            return f"Late fine for {days} days is ₹{fine}"
    return "Please enter number of late days."

def library_chatbot():
    print(" Library Assistant Chatbot ")

    while True:
        try:
            user_input = input("User: ")
            intent = detect_intent(user_input)

            if intent == "unknown":
                print("Chatbot: Sorry, I didn't understand.")

            elif "book" in user_input:
                print("Chatbot:", check_availability(user_input))

            elif intent == "fine_info":
                print("Chatbot:", calculate_fine(user_input))

            else:
                response = random.choice(library_data["intents"][intent]["responses"])
                print("Chatbot:", response)

            if intent == "goodbye":
                break

        except Exception:
            print("Chatbot: Invalid input. Try again.")

library_chatbot()
```
